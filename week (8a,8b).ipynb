```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.cluster import AgglomerativeClustering

np.random.seed(42)
X = np.array([[2.5, 0], [3, 3], [5, 8], [8, 5], [10, 6]]) 
linkage_data = linkage(X, method='ward', metric='euclidean')
dendrogram(linkage_data, labels=['A', 'B', 'C', 'D', 'E'])
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

x = np.array([[2, 5], [3, 3], [5, 8], [8, 5], [10, 6]])

linkage_matrix = linkage(x, method='complete', metric='euclidean')

dendrogram(linkage_matrix, labels=['A', 'B', 'C', 'D', 'E'])

plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Euclidean Distance')
plt.show()
```
